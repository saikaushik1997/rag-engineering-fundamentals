```python
# Convert the chunk_text into dense vectors
dense_embeddings = pc.inference.embed(
    model="llama-text-embed-v2",
    inputs=[d['chunk_text'] for d in data],
    parameters={"input_type": "passage", "truncate": "END"}
)

# Convert the same chunk_text into sparse vectors
sparse_embeddings = pc.inference.embed(
    model="pinecone-sparse-english-v0",
    inputs=[d['chunk_text'] for d in data],
    parameters={"input_type": "passage", "truncate": "END"}
)
```

Upsert one index with records of both dense and sparse vectors with different key/value pairs

```python
# Target the hybrid index
index_host = pc.describe_index(index_name).host
index = pc.Index(host=index_host)

# Each record contains an ID, a dense vector, a sparse vector, and the original text as metadata
records = []
for d, de, se in zip(data, dense_embeddings, sparse_embeddings):
    records.append({
        "id": d['_id'],
        "values": de['values'],
        "sparse_values": {'indices': se['sparse_indices'], 'values': se['sparse_values']},
        "metadata": {'text': d['chunk_text']}
    })

# Upsert the records into the hybrid index
index.upsert(
    vectors=records,
    namespace="example-namespace"
)
```

Searching


Because the search algo views both the vectors as one unit, there's no built in parameter to weigh the dense part of the query against its sparse part. So the search gets top n of dense matches, and top n of sparse matches. By introducing a new score(function of alpha), we control if we want to give dense matches more weightage or sparse ones. Depends on the nature of the data and the queries - keyword match(low alpha), semantic match(high alpha). 

```python
def hybrid_score_norm(dense, sparse, alpha: float):
    """Hybrid score using a combination

    alpha * dense value + (1 - alpha) * sparse value

    Args:
        dense: Array of floats representing
        sparse: a dict of `indices` and `values`
        alpha: scale between 0 and 1
    """
    if alpha < 0 or alpha > 1:
        raise ValueError("Alpha must be between 0 and 1")
    hs = {
        'indices': sparse['indices'],
        'values':  [v * (1 - alpha) for v in sparse['values']]
    }
    return [v * alpha for v in dense], hs
```

```python
# Define the query
query = "Unemployment stats"

# dense embedding of the query
dense_query = pc.inference.embed(
    model="llama-text-embed-v2",
    inputs=[query],
    parameters={"input_type": "query", "truncate": "END"}
)[0]['values']

# sparse embedding of the query
sparse_query = pc.inference.embed(
    model="pinecone-sparse-english-v0",
    inputs=[query],
    parameters={"input_type": "query", "truncate": "END"}
)[0]

sparse_query_fmt = {
    'indices': sparse_query['sparse_indices'],
    'values':  sparse_query['sparse_values']
}

# Apply hybrid normalization (0.7 = leaning toward semantic/dense)
alpha = 0.7
hdense, hsparse = hybrid_score_norm(dense_query, sparse_query_fmt, alpha)

# Search the dense index
results = index.query(
    namespace="example-namespace",
    vector=hdense,
    sparse_vector=hsparse,
    top_k=20,
    include_metadata=True
)

for match in results['matches']:
    print(f"[{match['score']:.4f}] {match['metadata']['text']}")

```

```python
# Cohere re-ranking
import cohere
co = cohere.ClientV2()

# results from search being passed to reranker
search_results=[match['metadata']['text'] for match in results['matches']]
rerank_results = co.rerank(
    model="rerank-v4.0-pro", query=query, documents=search_results, top_n=5,
)

print(rerank_results)

for result in rerank_results.results:
    print(f"Rank: {result.index}")
    print(f"Score: {result.relevance_score}")
    print(f"Doc: {search_results[result.index]}")
```
